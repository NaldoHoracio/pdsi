### Histograma de imagens

# Histograma

É um gráfico em 2D que mostra a distribuição de intensidade em cada pixel de uma imagem

- Segmentação
- Análise estatística

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("baboon.png", cv2.IMREAD_GRAYSCALE)
```

```python
img.shape
```

```python
hist_img = cv2.calcHist([img], [0], None, [255], [0, 255])

plt.figure(figsize=(5, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap="gray")
plt.title("Baboon")
```

```python
plt.subplot(1, 2, 2)
plt.plot(hist_img, color="blue")
plt.title("Histagrama do Baboon")
plt.ylabel("Número de Pixels")
plt.xlabel("Intensidade")
plt.show()
```

### Equalização de imagem

Técnica usada para ajustar o contraste de uma imagem (diferença de intensidade entre os pixels)

```python
img_equalizer = cv2.equalizeHist(img)

# @param img: lista de imagens
# @param channels: numero de canais
# @aparam mask: mascara usada
# @param bins: numero de bins (retangulos) para formar o histograma
# @param ranges: intervalo com valores maximos e minimos
hist_original = cv2.calcHist([img], [0], None, [255], [0, 255])
hist_equalizer = cv2.calcHist([img_equalizer], [0], None, [255], [0, 255])

plt.figure(figsize=(7, 6))

plt.subplot(2, 2, 1)
plt.plot(hist_original, color="black")
plt.title("Histograma Original")
plt.ylabel("Número de Pixels")
plt.xlabel("Intensidade")

plt.subplot(2, 2, 2)
plt.plot(hist_equalizer, color="blue")
plt.title("Histograma Equalizado")
plt.ylabel("Número de Pixels")
plt.xlabel("Intensidade")
```

```python
import numpy as np
from PIL import Image

img = Image.open("baboon.png").convert("L")
img_array = np.array(img)
```

### Negativo de uma imagem

O negativo de uma imagem é o valor de intensidade onde cada pixel tem seu valor subtraído em relação ao seu valor máximo.

```python
img_array = np.array(img)

negative = 255 - img_array
plt.imshow(negative, cmap="gray")
plt.title("Negativo da escala de cinza")
plt.show()
```

### Correção Gamma

Tipo de transformação (mudança) em uma imagem para modificar os valores de brilho dos pixels.

```python
import numpy as np

gamma = 2.2
gamma_corrected = np.array(255*(img_array/255) ** (1/gamma), dtype='uint8')
plt.imshow(gamma_corrected, cmap="gray")
plt.title("Correção Gama")
plt.show()
```

# Máscaras em imagens

É uma matriz de valores passada sobre uma imagem para modificar seus pixels de acordo com uma operação matemática

```python
# Exemplo: cria uma máscara para substituir todos os valores em vermelho por 255
mascara = img_array > 128

maskared_array = np.zeros(img_array.shape, dtype=np.uint8)
maskared_array[mascara] = 255

plt.imshow(maskared_array, cmap="gray")
plt.title("Máscara Vermelha")
plt.show()
```
